# Pickleball Hotspot Analysis
### A Data-Driven Strategy for Tournament Expansion

Pickleball is the fastest-growing sport in the US, but not every city is ready to host a major tournament. This project analyzes pickleball infrastructure, demand signals, and demographic context to uncover where the next major events should go and why one tournament didnâ€™t work out.

## Why This Matters
The US Open Pickleball Championships, held annually in Naples, Florida, is the most prestigious event in the sport. With growing demand other large tournaments, there's a need to identify other cities that offer the right mix of facilities, demand, and demographics.

This project began with a question: **Why did a tournament in Jersey City fail to gain traction?** That case helped shape the analysis pointing to patterns that predict both success and failure.

## Pipeline Overview
I pulled and merged data from five key sources:
- Court locations ([Pickleheads](https://www.pickleheads.com))
- Tournament activity ([PickleballTournaments.com](https://pickleballtournaments.com/))
- Google Trends search interest
- U.S. Census city demographics (population, income, age)
- A Random Forest model to estimate major tournament potential

Each city received a **composite score**, and I predicted its likelihood of hosting a major event using historical patterns.

```python
# Import necessary python libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as mtick
import seaborn as sns
```

```python
# Load and preview the final cleaned and merged dataset containing city-level pickleball data
df = pd.read_csv('clean-data/merged_city_data_with_composite_score.csv')
df.head()
```

## Composite Score Breakdown
Cities were scored using normalized values from:
- Total pickleball courts
- Total tournaments and large tournament counts
- Player counts
- Courts per 10,000 people

I then used a Random Forest model to estimate whether each city could realistically host a major tournament.

### Top 10 Cities by Composite Score

```python
# Select top 10 cities with the highest composite score
top = df.sort_values(by='composite_score', ascending=False).head(10)

# Create bar chart to visualize the top cities' composite scores
plt.figure(figsize=(12, 6))
bars = sns.barplot(data=top, x='composite_score', y='city_state', color='#4CAF50')

# Add labels to the bars
for i, v in enumerate(top['composite_score']):
    plt.text(v + 0.01, i, f'{v:.2f}', va='center')

# Add title and axis labels and display
plt.title('Top 10 Cities by Composite Score')
plt.xlabel('Composite Score')
plt.ylabel('City')
plt.tight_layout()
plt.show()
```

### High Potential Cities Without a Major Tournament

```python
# Select top 10 cities with the highest untapped composite score (have not yet hosted a major)
untapped = df[df['has_major_tournament'] == 0].sort_values(by='composite_score', ascending=False).head(10)

# Create bar chart to visualize the top cities' untapped composite scores
plt.figure(figsize=(12, 6))
bars = sns.barplot(data=untapped, x='composite_score', y='city_state', color='#4CAF50')

# Add labels to the bars
for i, v in enumerate(untapped['composite_score']):
    plt.text(v + 0.01, i, f'{v:.2f}', va='center')

# Add title and axis labels and display
plt.title('Top 10 Untapped Cities by Composite Score')
plt.xlabel('Composite Score')
plt.ylabel('City')
plt.tight_layout()
plt.show()
```

## Demographics and Composite Score

```python
# Map binary major tournament flag for plotting and labeling
df['tournament_label'] = df['has_major_tournament'].map({1: 'Has Major', 0: 'No Major'})

# Create a scatterplot to show the relationship between median income and composite score
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='median_income', y='composite_score', hue='tournament_label', palette='Greens', alpha=0.8)

# Add title and axis labels
plt.title('Composite Score vs Median Income')
plt.xlabel('Median Income')
plt.ylabel('Composite Score')

# Format x-axis labels as currency and display
plt.gca().xaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))
plt.tight_layout()
plt.show()
```

```python
# Create a scatterplot to show the relationship between median age and composite score
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='median_age', y='composite_score', hue='tournament_label', palette='Greens', alpha=0.8)

# Add title and axis labels and display
plt.title('Composite Score vs Median Age')
plt.xlabel('Median Age')
plt.ylabel('Composite Score')
plt.tight_layout()
plt.show()
```

## Tournament Seasonality

```python
# Load cleaned tournament data
tournaments = pd.read_csv('clean-data/tournaments_cleaned.csv')

# Convert start dates to datetime format and extract the month name
tournaments['Start Date'] = pd.to_datetime(tournaments['Start Date'], errors='coerce')
tournaments['Month'] = tournaments['Start Date'].dt.month_name()
ordered_months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Count tournaments by month and ensure correct order
month_counts = tournaments['Month'].value_counts().reindex(ordered_months)

# Create a line chart to visualize seasonality of tournament activity
plt.figure(figsize=(12, 8))
plt.plot(month_counts.index, month_counts.values, marker='o', color='#4CAF50')

# Add labels above each point
for i, value in enumerate(month_counts.values):
    plt.text(i, value + 5, str(value), ha='center')

# Add title and axis labels and display
plt.title('Number of Tournaments by Month (2023)')
plt.xlabel('Month')
plt.ylabel('Tournaments')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Jersey City Analysis: A Look Into What Caused the Tournament to Fail

Jersey City once hosted a tournament in Febuary of 2023 (The 2023 Atlantic City Pickleball Open) that proved to be unsuccessful and did not return for 2024. It failed to draw the expected demand, and this project helped explain why:

**Findings:**
- Very few courts in the city
- Tournament was scheduled in **February**, the lowest-interest month
- Google Trends interest was flat in winter
- Demographics didnâ€™t match successful host cities

### Google Search Interest for 'Pickleball Tournament' in 2023:

```python
# Load cleaned and merged Google Trends data and filter to 2023
trend = pd.read_csv('clean-data/merged_google_trend_time.csv')
trend['date'] = pd.to_datetime(trend['date'])
trend = trend[['date', 'pickleball_tournament']].dropna()
trend = trend[trend['date'].dt.year == 2023].rename(columns={'pickleball_tournament': 'interest'})


# Create a mask to highlight February (when Jersey City's tournament occurred)
feb_mask = trend['date'].dt.month == 2
line_color = '#4CAF50'

# Plot Google search interest in the term "Pickleball Tournament" over time in the year 2023
plt.figure(figsize=(11, 5))
plt.plot(trend['date'], trend['interest'], label='Search Interest', color=line_color)

# Highlight February with a shaded region to display the seasonal dip
plt.fill_between(trend['date'], trend['interest'], where=feb_mask, color='#fbeec1', alpha=0.4, label='February')

# Format with abbreviated month names
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())

# Add title and axis labels and display
plt.title("Google Trends (2023): 'Pickleball Tournament'")
plt.ylabel('Interest')
plt.xlabel('Date')
plt.legend()
plt.tight_layout()
plt.show()
```
